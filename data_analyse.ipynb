```python
# Replace the Null values in 'Review Rating' Column with median()
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.isna().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
# Rename the column name purchase_amount_(usd)

df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# Feauture Engineering
# Create a column age_groups

labels=['young Adult', 'Adult', 'Middle-aged', 'Senior']

df['age_groups'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_groups']].head(10)
```

```python
# Create a column for purchase_frequency_days with help of this column 'frequency_of_purchases'

frequency_mapping = {
    'Fortnightly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```
